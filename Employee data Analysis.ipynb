## Dataset :- Human Resources Data Set (kaggle)

### Exploratory data analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("HRDataset_v14.csv")
```

```python
data.head()
```

#### checking DataFrame

```python

print(data.info())
```

#### checking missing values

```python

data.isnull().sum()
```

##### null only in date of termination

### statistical measure of dataset

```python

data.describe()
```

### checking correlation between columns

```python

plt.figure(figsize=(15,10))
sns.heatmap(data.corr(),annot=True)
plt.show()
```

### checking gender distribution

```python
sns.countplot(x=data['Sex'])
```

#### there are more females then men in the company 

### checking performance distribution 

```python

plt.figure(figsize=(10,5))
sns.catplot(x='PerformanceScore', data=data, kind="count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='PerformanceScore', data=data, hue = 'Sex')
plt.show()
```

#### performnace distribution is fairly even across gender

```python
plt.figure(figsize=(10,5))
sns.countplot(x='PerformanceScore', data=data, hue = 'Termd')
plt.show()
```

#### people who were terminated did not meet performance expectation 

```python
plt.figure(figsize=(10,5))
sns.countplot(x='PerformanceScore', data=data, hue = 'Department')
plt.show()
```

###  various recruitment source

```python

data['RecruitmentSource'].unique()
```

```python
sns.countplot(y='RecruitmentSource', data=data )
plt.show()
```

#### most of the employee are hired through linkedin and indeed

### salary distribution

```python

sns.histplot(data['Salary'])
```

### avg salary by department

```python

avg_sal = data.groupby('Department')['Salary'].mean()
avg_sal
```
